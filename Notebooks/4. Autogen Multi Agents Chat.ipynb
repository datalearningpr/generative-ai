# Autogen Multi Agents Chat

This notebook illustrates utilizing the Autogen library to efficiently configure various conversational scenarios in which distinct agents assume different roles. The entire chat process is automated. Autogen use cases primarily involve resolving issues related to coding or tools. The examples here aim to demonstrate that by establishing background and roles for different agents, the conversation itself can provide end-users with significant insights.




```python
import autogen

config_list_gpt4 = [

]
```

### Example 1: Debates

```python
llm_config = {"config_list": config_list_gpt4, "seed": 42}
user_proxy = autogen.UserProxyAgent(
   name="User_proxy",
   system_message="A human admin.",
   human_input_mode="TERMINATE"
)
d1 = autogen.AssistantAgent(
    name="debater1",
    system_message="Strong debater in defending its own opinion.",
    llm_config=llm_config,
)
d2 = autogen.AssistantAgent(
    name="debater2",
    system_message="Strong debater in defending its own opinion.",
    llm_config=llm_config,
)
groupchat = autogen.GroupChat(agents=[user_proxy, d1, d2], messages=[], max_round=12)
manager = autogen.GroupChatManager(groupchat=groupchat, llm_config=llm_config)


user_proxy.initiate_chat(manager, message="""
Which is More Important: Hard Work or Talent?
debater1 think it is hard work.
debater2 think it is talent.
start the debate.
""")
```

### Example 2: Content Creation

```python
llm_config = {"config_list": config_list_gpt4, "seed": 42}
user_proxy = autogen.UserProxyAgent(
   name="User_proxy",
   system_message="A human admin.",
   human_input_mode="TERMINATE"
)
d1 = autogen.AssistantAgent(
    name="teacher",
    system_message="""
    Excellent teacher who will teach new technology to non IT major students in university
    Each course will have a topic which will last a 3 hours session.
    Given a selected topic, you will give out the detailed course outline and materials and how the course will be carried out.
    """,
    llm_config=llm_config,
)
d2 = autogen.AssistantAgent(
    name="critic",
    system_message="""
    Excellent course critic who will give comments and suggestion for an given course outline, material and teaching method.
    """,
    llm_config=llm_config,
)
groupchat = autogen.GroupChat(agents=[user_proxy, d1, d2], messages=[], max_round=12)
manager = autogen.GroupChatManager(groupchat=groupchat, llm_config=llm_config)

user_proxy.initiate_chat(manager, message="""
we need a new course next month for topic "Javascript"
""")
```
